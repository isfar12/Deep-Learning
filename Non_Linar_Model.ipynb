````python
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model

# Input layer
inputs = Input(shape=(10,))

# Layers as functions
x = Dense(32, activation='relu')(inputs)
x = Dense(16, activation='relu')(x)
output1 = Dense(1, activation='sigmoid')(x)
output2 = Dense(1, activation='sigmoid')(x)

# Define model
model = Model(inputs=inputs, outputs=[output1,output2])

model.summary()
````

````python
from keras.utils import plot_model
plot_model(model,show_shapes=True)
````

`plot_model()` from Keras requires **Graphviz** — both the **Python package** *and* the **system-level tool** to be installed properly.

Let me guide you step-by-step to fix it 💡

---

### ✅ Step 1: Install Python Graphviz library

You need this for Python to *talk* to Graphviz:

```bash
pip install graphviz
```

---

### ✅ Step 2: Install the **Graphviz system package** (This is the part most people miss)

This actually installs the `.dot` engine that draws the diagrams. Here's how:

#### 🪟 **On Windows:**
1. Go to: https://graphviz.gitlab.io/download/
2. Download the **Graphviz Windows installer** (usually `.exe`)
3. Install it (default options are fine)
4. Add Graphviz to the **system PATH**:
   - Search "Environment Variables" in Start Menu
   - Under **System variables → Path → Edit**
   - Add the path where Graphviz is installed, usually:
     ```
     C:\Program Files\Graphviz\bin
     ```

### ✅ Step 3: Use `plot_model()`

```python
from tensorflow.keras.utils import plot_model

plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

Then open `model_plot.png` to see your DAG-style architecture.
