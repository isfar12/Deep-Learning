```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("../Dataset/SalesData.csv")
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
x=df[["Temperature"]]
y=df["Revenue"]
```

```python
sns.scatterplot(x=x.Temperature,y=y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=0)
```

```python
x_train
```

```python
model=tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=1,input_shape=[1]))
```

```python
model.summary()
```

```python
model.compile(loss="mse",optimizer=tf.keras.optimizers.Adam(0.1))
```

```python
history=model.fit(x_train,y_train,epochs=500)
```

```python
history.history
```

```python
history.history.keys()
```

```python
plt.plot(history.history["loss"])
```

```python
scaler=StandardScaler()
x_train_scaler=scaler.fit_transform(x_train)
x_test_scaler=scaler.transform(x_test)
```

```python
model2=tf.keras.Sequential()
model2.add(tf.keras.layers.Dense(1,input_shape=(1,)))
```

```python
model2.summary()
```

```python
history2=model2.compile(loss="mse",optimizer=tf.keras.optimizers.Adam(0.1))
```

```python
history2=model2.fit(x_train_scaler,y_train,epochs=700)
```

```python
plt.plot(history2.history["loss"])
```

```python
model2.get_weights()
```

```python
linear_model=LinearRegression()
linear_model.fit(x_train,y_train)
```

```python
y_pred_ann=model.predict(x_test)
```

```python
plt.plot(x_test,y_test)
plt.plot(x_test,y_pred_ann,"r")
plt.legend(["Y Test","Y Predict"])
plt.show()
```

```python
y_pred_ann2=model2.predict(x_test_scaler)
```

```python
plt.plot(x_test,y_test)
plt.plot(x_test,y_pred_ann2,"r.")
plt.legend(["Y Test","Y Predict"])
```

```python
y_pred_ml=linear_model.predict(x_test)
```

```python
plt.plot(x_test,y_test)
plt.plot(x_test,y_pred_ml,"r")
plt.legend(["Y Test","Y Predict"])
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred_ann)
```
